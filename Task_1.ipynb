### Load and Clean Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "Dataset\\API_SP.POP.TOTL_DS2_en_excel_v2_130162.xlsx"
excel_data = pd.ExcelFile(file_path)

df = pd.read_excel(file_path, sheet_name="Data")

df_clean = df[3:]
df_clean.columns = df.iloc[2]
df_clean = df_clean.dropna(subset=["Country Name"])

year_columns = df_clean.columns[4:]
df_clean[year_columns] = df_clean[year_columns].apply(pd.to_numeric, errors="coerce")

top_10_2020 = df_clean[["Country Name", 2020.0]].nlargest(10, 2020.0)
```

### Bar Chart

```python
plt.figure(figsize=(10, 6))
plt.bar(top_10_2020["Country Name"], top_10_2020[2020.0], color='skyblue')
plt.title("Top 10 Most Populous Countries in 2020")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Histogram

```python
plt.figure(figsize=(10, 6))
plt.hist(df_clean[2020.0].dropna(), bins=30, color='lightgreen', edgecolor='black')
plt.title("Distribution of Country Populations (2020)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
